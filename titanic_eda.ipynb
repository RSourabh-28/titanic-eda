```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(df)
```

```python
df_copy = df.copy()
```

```python
print(df_copy.shape)
print(df_copy.info())
print(df_copy.describe())
```

```python
print(df_copy.isnull().sum())
```

```python
print(df_copy.isnull().mean())
```

```python
df_copy["Age"].fillna(df_copy["Age"].median(), inplace=True)
```

```python
df_copy["Embarked"].fillna(df_copy["Embarked"].mode()[0], inplace=True)
```

```python
df_copy.drop(columns = ["Cabin"], inplace=True)
```

```python
df_copy["Age"] = df_copy["Age"].astype(int)
```

```python
plt.Figure(figsize=(8,11))
sns.histplot(df_copy["Age"], bins=20)
plt.title("Age Distribution", fontsize=12)
plt.xlabel("Age",fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(8,11))
sns.heatmap(df_copy.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation of data")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x = "fare", y = "" , data = df_copy)
plt.title("Fare Vs survival")
plt.xlabel("fare",fontsize=12)
plt.ylabel("Fare",fontsize=12)
plt.tight_layout()
plt.show()
```
